```python
#Importing packages
import pandas as pd
import seaborn as sns
```

```python
#Read .csv
hosts = pd.read_csv("olympic_hosts.csv")
hosts
```

```python
athletes = pd.read_csv("olympic_athletes.csv")
athletes
```

```python
medals = pd.read_csv("olympic_medals.csv")
medals
```

```python
results = pd.read_csv("olympic_results.csv")
results
```

```python
data = medals.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
data
```

```python
data = data.loc[(data["country_name"]=="Italy") & (data["game_season"]=="Summer")]
data
```

```python
#Use the function groupby for having the number of athletes
#in each discipline
a = data.groupby(by = ['discipline_title']).size().reset_index(name = 'athletes')
a
```

```python
#Number of athletes in each year
b = data.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
b
```

```python
idx = a["discipline_title"]
idx
```

```python
cols = b["game_year"]
cols
```
