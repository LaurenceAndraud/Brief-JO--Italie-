```python
#Importing packages
import pandas as pd
import seaborn as sns
```

```python
#Read .csv
hosts = pd.read_csv("olympic_hosts.csv")
hosts
```

```python
athletes = pd.read_csv("olympic_athletes.csv")
athletes
```

```python
medals = pd.read_csv("olympic_medals.csv")
medals
```

```python
data = medals.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
data
```

```python
data = data.loc[(data["country_name"]=="Italy") & (data["game_season"]=="Summer")]
data
```

```python
data.info()
```

```python
#Use the function .values for seeing some names of athletes
#and confirm if some names appaers various time
p = data["athlete_full_name"].values
p
```

```python
#Associate a name with his number of medals
c = data.groupby(by = ['athlete_full_name']).size().reset_index(name = 'medals_type')
c
```

```python
#Select a top 10 entries with the best number of medals
t = c['medals_type'].nlargest(n=10)
t
```
