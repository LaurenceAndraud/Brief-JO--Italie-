```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympic_medals = pd.read_csv( 'olympic_medals.csv')
olympic_hosts = pd.read_csv( 'olympic_hosts.csv')
olympic_medals
```

```python
# Je veux les médailles et l'italie
italy_medals = olympic_medals.loc[(olympic_medals['country_name']=='Italy')]
italy_medals
```

CONSIGNE : une courbe qui représente le nombre de médailles gagnées au cours du temps

Donc ma courbe à les années en abscisse et le nbs en ordonnée

finalement, je vais faire une df avec les jeux d'été et l'année

```python
olympic_hosts
```

```python
# filter par jeux d'été
summer_table = olympic_hosts.loc[(olympic_hosts['game_season']=='Summer')]
summer_table
```

```python
# merger la table médailles et la table host via les colonnes sulg game et game slug
data = olympic_medals.merge(olympic_hosts, how='left', left_on='slug_game', right_on='game_slug')
data
```

```python
italy = data.loc[(data['country_name']=='Italy')]
```

```python
summer= italy.loc[(italy['game_season']=='Summer')]
summer
```

```python
result = summer.groupby(by = ['game_year']).size().reset_index(name='medals')
# data
# size-> avoir toute la colonne
# reset_index -> pour stocker les valeurs de la nouvelle colonne 
result
```

```python
sns.lineplot(data = result , x="game_year", y="medals").set_title("Médailles gagnées au cours des années par l'Italie")
plt.savefig('medailles')
```
