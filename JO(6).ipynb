```python
#Importing packages
import pandas as pd
import seaborn as sns
```

```python
#Read .csv
hosts = pd.read_csv("olympic_hosts.csv")
hosts
```

```python
athletes = pd.read_csv("olympic_athletes.csv")
athletes
```

```python
medals = pd.read_csv("olympic_medals.csv")
medals
```

```python
results = pd.read_csv("olympic_results.csv")
results
```

```python
data = medals.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
data
```

```python
data = data.loc[(data["country_name"]=="Italy") & (data["game_season"]=="Summer")]
data
```

```python
#nbre de médailles gagnées
data = data.groupby(by = ['game_year']).size().reset_index(name = 'medal_type')
data
```

```python
a = results.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
a
```

```python
b = a.loc[(a["country_name"]=="Italy") & (a["game_season"]=="Summer")]
b
```

```python
#nbre d'athlètes engagés
c = b.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
c
```

```python
#merging for having informations about number of athletes & number of medals in the same array
e = c.merge(data, on="game_year", how='outer', indicator=True)
e
```

```python
sns.scatterplot(data=e, x='athletes', y='medal_type').set_title('Corrélation entre le nombre de participants et le nombre de médailles')
```
