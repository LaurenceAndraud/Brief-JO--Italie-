```python
#Importing packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read .csv
hosts = pd.read_csv("olympic_hosts.csv")
hosts
```

```python
athletes = pd.read_csv("olympic_athletes.csv")
athletes
```

```python
#Seeing names of columns and the count
athletes.info()
```

```python
medals = pd.read_csv("olympic_medals.csv")
medals
```

```python
results = pd.read_csv("olympic_results.csv")
results
```

```python
#Merging two files.csv on a very similar column
data = medals.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
data
```

```python
#Filter the dataframe for having only Italians, and Summer session
data = data.loc[(data["country_name"]=="Italy") & (data["game_season"]=="Summer")]
data
```

```python
p = data["event_gender"].values
p
```

```python
a = results.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
a
```

```python
b = a.loc[(a["country_name"]=="Italy") & (a["game_season"]=="Summer")]
b
```

```python
b.info()
```

```python
#Number of participant for each year
c = b.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
c
```

```python
men = data.loc[(data["event_gender"]=="Men")]
men
```

```python
men.info()
```

```python
#Number of men participants per years
d = men.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
d
```

```python
#Number of woman participants per years
woman = data.loc[(data["event_gender"]=="Women")]
woman
```

```python
e = woman.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
e
```

```python
mix = data.loc[(data["event_gender"]=="Mixed")]
mix
```

```python
g = mix.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
g
```

```python
ope = data.loc[(data["event_gender"]=="Open")]
ope
```

```python
h = ope.groupby(by = ['game_year']).size().reset_index(name = 'athletes')
h
```

```python
#Figure for seeing Italians participants per years and per gender(Men/Woman) (2)
sns.lineplot(data = c, x = "game_year", y = "athletes", legend=True).set_title("Evolution du nombre de participants")
sns.lineplot(data = d, x = "game_year", y = "athletes", legend=True)
sns.lineplot(data = e, x = "game_year", y = "athletes", legend=True)
#sns.lineplot(data = g, x = "game_year", y = "athletes") #données trop petites, à tenir en compte ?
#sns.lineplot(data = h, x = "game_year", y = "athletes") #données trop petites, et pas sûre de savoir ce à quoi correspond 'open'
plt.legend(labels=["Total","Homme","Femme"])
```
