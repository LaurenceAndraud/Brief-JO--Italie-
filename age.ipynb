```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
year_birth = pd.read_csv('olympic_athletes.csv')
general = pd.read_csv('it_summer.csv')
year_birth
```

```python
# je lie la table des athlètes sur leur url

result = pd.merge(general, year_birth, left_on="athlete_url", right_on="athlete_url")
result

# https://www.pythoniste.fr/python/pandas/fusionner-des-dataframes-a-laide-de-la-methode-merge/
# meilleur lien pour apprendre les merge!!
```

```python
# maintenant je soustrait la colone game_year à la colonne year birth et stocker le résultat dans uen nouvelle colone 
result = result.assign(age = result['game_year'] - result['athlete_year_birth'])
result
```

```python
# certaines lignes de la colonne age sont nulles je vais regarder combien et les enlever au besoin
result.isnull().sum()
# il y a deux valeurs nulles dans âge, sur 481 valeurs, je peux me permettre de les enlever 
```

```python
# enlever les valeurs nulles de âge 
# ww =result.drop(result.loc[result['age']=='NaN'].index, inplace=True)

result.drop(result[result['age'].isnull()].index, inplace=True)
# inplace permet de restocker dans la variable actuelle
```

```python
result
```

```python
# cette commande permet d'appeler les colonnes d'un même dataframe 
df = result[['age','game_year']]
df
```

```python
sns.histplot(data = df, x = 'age').set_title("Age des participants sur la globalité des éditions aux JO pour l'Italie")
plt.savefig('age')
```
