```python
#Importing packages
import pandas as pd
import seaborn as sns
```

```python
#Read .csv
hosts = pd.read_csv("olympic_hosts.csv")
hosts
```

```python
athletes = pd.read_csv("olympic_athletes.csv")
athletes
```

```python
#Seeing names of columns and the count
athletes.info()
```

```python
medals = pd.read_csv("olympic_medals.csv")
medals
```

```python
medals.info()
```

```python
#Merging two files.csv on a very similar column
data = medals.merge(hosts, right_on="game_slug", left_on="slug_game", how='outer', indicator=True)
data
```

```python
#Filter the dataframe for having only Italians, and Summer session
data = data.loc[(data["country_name"]=="Italy") & (data["game_season"]=="Summer")]
data
```

```python
#Use groupby for having medals per years
data = data.groupby(by = ['game_year']).size().reset_index(name = 'medals')
data
```

```python
#Figure for seeing Italians medals per years (1)
sns.lineplot(data = data, x = "game_year", y = "medals").set_title("Evolution du nombre de médailles gagnées")
```
